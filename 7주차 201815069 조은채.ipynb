# Subplots

```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
plt.figure()
plt.plot(x, '-o')
```

```python
plt.figure()
plt.subplot(1, 2, 1)   #행, 열, 인덱스 => 첫번째 열만 사용하겠다는 것

linear_data = x
plt.plot(linear_data, '-o')
```

```python
exponential_data = linear_data**2

plt.subplot(1, 2, 2)
plt.plot(exponential_data, '-o')
```

```python
plt.subplot(1, 2, 1)
plt.plot(exponential_data, '-o')
```

```python
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, '-o')

ax2 = plt.subplot(1, 2, 2, sharey = ax1)
plt.plot(exponential_data, '-x')
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex = True, sharey = True)
```

```python
ax5.plot(linear_data, '--')
```

```python
ax7.plot(linear_data, '-x')
```

```python
plt.figure()
plt.subplot(1, 2, 1) == plt.subplot(121)
```

# Histograms

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]
```

```python
for n in range(0, len(axs)):
    sample_size = 10**(n + 1)
    sample = np.random.normal(loc = 0.0, scale = 1.0, size = sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n = {}'.format(sample_size))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n + 1)
    sample = np.random.normal(loc = 0.0, scale = 1.0, size = sample_size)
    axs[n].hist(sample, bins = 100)    #bins == 구간 => 100마다 구간 자른 것
    axs[n].set_title('n = {}'.format(sample_size))
```

```python
plt.figure()
Y = np.random.normal(loc = 0.0, scale = 1.0, size = 10000) #loc = 평균, scale = 분산
X = np.random.random(size = 10000)
plt.scatter(X, Y)
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0,1:])    #첫번째 row에서 두번째컬럼부터 쭉 사용하겠다는 뜻
side_histogram = plt.subplot(gspec[1:,0])   #두번째 로우에서 세번째로우, 첫번째 컬럼에 그리겠다
lower_right = plt.subplot(gspec[1:,1:])     #두번째 로우 세번째로우, 두번째컬럼 세번째 컬럼에 그리겠다
```

```python
lower_right.scatter(X, Y)
top_histogram.hist(X, bins = 100)
s = side_histogram.hist(Y, bins = 100, orientation = 'horizontal')   #vertical로 있던것을 horizontal로
```

```python
top_histogram.clear()    #그래프 지우는 것
top_histogram.hist(X, bins = 100, density = True)
side_histogram.clear()
side_histogram.hist(Y, bins = 100, orientation = 'horizontal', density = True)
side_histogram.invert_xaxis()    #좌우반전
```
