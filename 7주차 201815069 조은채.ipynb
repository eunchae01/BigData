# Subplots

```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
plt.figure()
plt.plot(x, '-o')
```

```python
plt.figure()
plt.subplot(1, 2, 1)   #í–‰, ì—´, ì¸ë±ìŠ¤ => ì²«ë²ˆì§¸ ì—´ë§Œ ì‚¬ìš©í•˜ê² ë‹¤ëŠ” ê²ƒ

linear_data = x
plt.plot(linear_data, '-o')
```

```python
exponential_data = linear_data**2

plt.subplot(1, 2, 2)
plt.plot(exponential_data, '-o')
```

```python
plt.subplot(1, 2, 1)
plt.plot(exponential_data, '-o')
```

```python
plt.figure()
ax1 = plt.subplot(1, 2, 1)
plt.plot(linear_data, '-o')

ax2 = plt.subplot(1, 2, 2, sharey = ax1)
plt.plot(exponential_data, '-x')
```

```python
fig, ((ax1, ax2, ax3), (ax4, ax5, ax6), (ax7, ax8, ax9)) = plt.subplots(3, 3, sharex = True, sharey = True)
```

```python
ax5.plot(linear_data, '--')
```

```python
ax7.plot(linear_data, '-x')
```

# Histograms

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]
```

```python
for n in range(0, len(axs)):
    sample_size = 10**(n + 1)
    sample = np.random.normal(loc = 0.0, scale = 1.0, size = sample_size)
    axs[n].hist(sample)
    axs[n].set_title('n = {}'.format(sample_size))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex = True)
axs = [ax1, ax2, ax3, ax4]

for n in range(0, len(axs)):
    sample_size = 10**(n + 1)
    sample = np.random.normal(loc = 0.0, scale = 1.0, size = sample_size)
    axs[n].hist(sample, bins = 100)    #bins == êµ¬ê°„ => 100ë§ˆë‹¤ êµ¬ê°„ ìžë¥¸ ê²ƒ
    axs[n].set_title('n = {}'.format(sample_size))
```

```python
plt.figure()
Y = np.random.normal(loc = 0.0, scale = 1.0, size = 10000) #loc = í‰ê· , scale = ë¶„ì‚°
X = np.random.random(size = 10000)
plt.scatter(X, Y)
```

```python
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0,1:])    #ì²«ë²ˆì§¸ rowì—ì„œ ë‘ë²ˆì§¸ì»¬ëŸ¼ë¶€í„° ì­‰ ì‚¬ìš©í•˜ê² ë‹¤ëŠ” ëœ»
side_histogram = plt.subplot(gspec[1:,0])   #ë‘ë²ˆì§¸ ë¡œìš°ì—ì„œ ì„¸ë²ˆì§¸ë¡œìš°, ì²«ë²ˆì§¸ ì»¬ëŸ¼ì— ê·¸ë¦¬ê² ë‹¤
lower_right = plt.subplot(gspec[1:,1:])     #ë‘ë²ˆì§¸ ë¡œìš° ì„¸ë²ˆì§¸ë¡œìš°, ë‘ë²ˆì§¸ì»¬ëŸ¼ ì„¸ë²ˆì§¸ ì»¬ëŸ¼ì— ê·¸ë¦¬ê² ë‹¤
```

```python
lower_right.scatter(X, Y)
top_histogram.hist(X, bins = 100)
s = side_histogram.hist(Y, bins = 100, orientation = 'horizontal')   #verticalë¡œ ìžˆë˜ê²ƒì„ horizontalë¡œ
```

```python
top_histogram.clear()    #ê·¸ëž˜í”„ ì§€ìš°ëŠ” ê²ƒ
top_histogram.hist(X, bins = 100, density = True)
side_histogram.clear()
side_histogram.hist(Y, bins = 100, orientation = 'horizontal', density = True)
side_histogram.invert_xaxis()    #ì¢Œìš°ë°˜ì „
```
