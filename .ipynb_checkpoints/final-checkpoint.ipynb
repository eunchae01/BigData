# Machine Learining

# K-NN classification

선호하는 프로그램 찾기

K-NN classification으로 프로그램의 방송연도와 평점 성인방송인지의 조건을 넣어 그 조건과 가장 비슷한 프로그램을 찾도록 하겠습니다.

```python
lookup_title = dict(zip(netflix.orign_country.unique(), netflix.title.unique()))
```

```python
lookup_title
```

```python
X = netflix[['startYear', 'endYear', 'rating', 'isAdult']]
y = netflix['orign_country']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```

```python
X_train.head()
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

```python
knn.score(X_train, y_train)
```

1)2000년에 시작해서 2015년에 끝나고 평점은 8인 방송

```python
netflix_prediction = knn.predict([[2000, 2015, 8, 0]])
lookup_title[netflix_prediction[0]]
```

2)2018년에 시작해서 2020년에 끝내고 평점은 5.5인 성인방송

```python
netflix_prediction = knn.predict([[2018, 2020, 5.5, 1]])
lookup_title[netflix_prediction[0]]
```

3)2010년에 시작해서 2011년에 끝나고 평점은 4인 방송

```python
netflix_prediction = knn.predict([[2010, 2011, 4, 0]])
lookup_title[netflix_prediction[0]]
```
