```python
import pandas as pd
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

plt.figure()
gspec = gridspec.GridSpec(3, 3)

top_histogram = plt.subplot(gspec[0,:2])
side_histogram = plt.subplot(gspec[1:,:2])
lower_right = plt.subplot(gspec[1:,2:])

Y = np.random.normal(loc = 0.0, scale = 1.0, size = 10000) #loc = 평균, scale = 분산
X = np.random.normal(loc = 0.0, scale = 1.0, size = 10000)

top_histogram.hist(X, bins = 100, color = 'blue')
side_histogram.scatter(X, Y, color = 'red')
lower_right.hist(Y, bins = 100, orientation = 'horizontal', color = 'blue')
```
